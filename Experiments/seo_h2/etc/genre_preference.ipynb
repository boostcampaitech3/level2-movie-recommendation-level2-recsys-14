```python
import pandas as pd
import numpy as np
import os
from collections import deque
```

```python
genre_data= pd.read_csv(os.path.join('./data/train/genres.tsv'), sep='\t')
raw_data= pd.read_csv(os.path.join('./data/train/train_ratings.csv'), sep='\t')
```

```python
genre_df= pd.merge(raw_data, genre_data, on='item', how='left')
temp=pd.DataFrame(genre_df.groupby(['user','genre'])['item'].count())
temp.reset_index(inplace=True)
temp.sort_values(['user','item'],ascending=False, inplace=True)

temp2= temp.drop_duplicates('user', keep='first') # 유저별 가장 선호하는 장르

temp3= pd.merge(temp2, temp, on='user', how='left')
temp3.drop('item_x', axis=1, inplace=True)

# 가장 선호하는 장르와 그외 다른 장르들을 그루핑한다. 
temp3= pd.DataFrame(temp3.groupby(['genre_x','genre_y'])['item_y'].sum())
temp3.reset_index(inplace=True)
temp_pivot= temp3.pivot_table('item_y','genre_x','genre_y',margins=True)
# 각 행 별(가장 선호하는 장르 별) 비율로 값 수정
for i in range(len(temp_pivot)):
    temp_pivot.iloc[i]= temp_pivot.iloc[i] / temp_pivot.iloc[i,-1]

temp_pivot.drop(['All'], inplace=True)
temp_pivot.drop(['All'], axis=1, inplace=True)
temp_pivot.fillna(0, inplace=True)

del temp
del temp3
del genre_df
```
